```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dxdt(x,t):
    k=-0.22
    return k*x
t=np.linspace(0, 100 , 100)

solve_1=odeint(dxdt,y0=1,t=t)
solve_1=solve_1.flatten()

def euler(f, y0=1, H=0.01, n = 100):
    eul = np.arange(0,n+H,H)
    y = [y0]
    for i in range(len(eul)-1):
        y.append(y[i] + H*f(y[i],eul[i]))
    return y, eul
solve_2, eul = euler(dxdt)

plt.plot(eul, solve_2, 'r', label = r"Метод Эйлера", color = "yellow")
plt.plot(t, solve_1, ':b', label=r"x'=kx", color = "red")
plt.grid(True)
plt.legend()
plt.show()


```
